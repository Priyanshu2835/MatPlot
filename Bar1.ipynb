```python
import matplotlib.pyplot as plt
import numpy as np
x=[2010,2011,2012,2013,2014,2015]
y=[4.11,4.89,6.23,7.98,11.89,12.88]
z=[3.12,4.12,6.89,7.93,11.76,13.97]
width=0.3
p1=[j+width for j in p]
p=np.arange(len(x))
plt.xlabel("Year",fontsize=15)
plt.ylabel("Revenue in USD",fontsize=15)
plt.title("Comp1 vs Comp2",fontsize=20)
plt.bar(p,y,width,color="r")
plt.bar(p1,z,width,color="y")
plt.show()
```
